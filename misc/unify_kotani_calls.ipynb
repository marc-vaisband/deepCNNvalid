```python
import numpy as np
import pandas as pd
```

```python
dtype_dict = {"Start": int, "End": int, "Chromosome": str}
kotani_snps = pd.read_csv("Supp table 3_SNP list_2019_Kotani.txt", skiprows=1, sep=" ", dtype=dtype_dict)
```

```python
kotani_snps
```

```python
kotani_snps["GL_ID"] = kotani_snps["Mouse"].apply(lambda x: f"{x}_tail")
kotani_snps["CL_ID"] = kotani_snps["Mouse"].apply(lambda x: f"{x}_BM")
kotani_snps["Func.RefGene"] = 0
```

```python
kotani_snps = kotani_snps.rename({"Chromosome": "chr", "Start": "start", "End": "end", "Reference": "ref", "Alteration": "alt", "Gene": "gene"}, axis=1)
```

```python
kotani_snps
```

```python
franz_checked_loci = pd.read_csv("./2021-10-13_testset1_nonsyn_fps_checked_Franz.csv", sep=";")
```

```python
franz_checked_loci
```

```python
franz_snps = franz_checked_loci[franz_checked_loci["comment"] == "TP"]
```

```python
franz_snps["Func.RefGene"] = 0
```

```python
franz_snps
```

```python
unified_cols = ["Mouse", "GL_ID", "CL_ID", "chr", "start", "end", "ref", "alt", "gene", "Func.RefGene"]
unified_table = pd.DataFrame(columns=unified_cols)

for i in range(len(kotani_snps)):
    unified_table = unified_table.append(kotani_snps.iloc[i][unified_cols])
for j in range(len(franz_snps)):
    unified_table = unified_table.append(franz_snps.iloc[j][unified_cols])
```

```python
unified_table.to_csv("kotani_snplist.csv", sep=";", header=True)
```
